**Fast Fourier Transform (FFT)**

```python
#Preprocessing FFT
from sklearn import preprocessing
from numpy.fft import rfft, rfftfreq

def FFT(data):
    data = np.asarray(data)
    #FFT
    n=len(data)
    dt=1/48000
    data=rfft(data)*dt
    freq=rfftfreq(n,dt)
    data=abs(data).T
    data = (np.delete(data, range(500*5,len(data)), axis=0)).T
    return (data)
```

**Feature Extraction**

```python
#Function Ekstraksi Fitur Statistik
import scipy
from scipy.stats import kurtosis
from scipy.stats import skew

def std(data):
    data = np.asarray(data)
    stdev=pd.DataFrame(np.std(data, axis=1))
    return stdev
def mean(data):
    data = np.asarray(data)
    M=pd.DataFrame(np.mean(data, axis=1))
    return M
def pp(data):
    data = np.asarray(data)
    PP=pd.DataFrame(np.max(data, axis=1) - np.min(data, axis=1))
    return PP
def Variance(data):
    data = np.asarray(data)
    Var=pd.DataFrame(np.var(data, axis=1))
    return Var
def rms(data):
    data = np.asarray(data)
    Rms=pd.DataFrame(np.sqrt(np.mean(data**2, axis=1)))
    return Rms
def Ab_mean(data):
    data = np.asarray(data)
    Abm=pd.DataFrame(np.mean(np.absolute(data),axis=1))
    return Abm
def Shapef(data):
    data = np.asarray(data)
    shapef=pd.DataFrame(rms(data)/Ab_mean(data))
    return shapef
def Impulsef(data):
    data = np.asarray(data)
    impulse=pd.DataFrame(np.max(data)/Ab_mean(data))
    return impulse
def crestf(data):
    data = np.asarray(data)
    crest=pd.DataFrame(np.max(data)/rms(data))
    return crest
def SQRT_AMPL(data):
    data = np.asarray(data)
    SQRTA = pd.DataFrame(np.sqrt(np.mean(np.absolute(data)**2, axis=1)))
    return SQRTA
def clearancef(data):
    data = np.asarray(data)
    max_amplitude = np.max(data, axis=1)
    sqrt_amplitude = SQRT_AMPL(data)
    clrf = pd.DataFrame(max_amplitude / sqrt_amplitude[0])
    return clrf
def kurtosis(data):
    data = pd.DataFrame(data)
    kurt = data.kurt(axis=1)
    return kurt
def skew(data):
    data = pd.DataFrame(data)
    skw = data.skew(axis=1)
    return skw
def ff(data):
    data = np.asarray(data)
    rms_val = np.sqrt(np.mean(data**2, axis=1))
    mean_val = np.mean(data, axis=1)
    form_factor_val = pd.DataFrame(rms_val / mean_val)
    return form_factor_val
```

```python
faults = ['BA_7', 'IR_7', 'OR1_7', 'OR2_7', 'OR3_7', 'BA_14', 'IR_14', 'OR1_14', 'BA_21', 'IR_21', 'OR1_21', 'OR2_21', 'OR3_21', 'N']

#Shape Factor
shapef_list = []
for fault in faults:
    shapef_df = Shapef(eval(fault))
    shapef_df['Fault'] = fault
    shapef_list.append(shapef_df)
shapef_all = pd.concat(shapef_list, axis=0, ignore_index=True)

#Root Mean Square (RMS)
rms_list = []
for fault in faults:
    rms_df = rms(eval(fault))
    rms_df['Fault'] = fault
    rms_list.append(rms_df)
rms_all = pd.concat(rms_list, axis=0, ignore_index=True)

#Peak to Peak
pp_list = []
for fault in faults:
    pp_df = pp(eval(fault))
    pp_df['Fault'] = fault
    pp_list.append(pp_df)
pp_all = pd.concat(pp_list, axis=0, ignore_index=True)

#Kurtosis
kurtosis_list = []
for fault in faults:
    data = eval(fault)
    kurtosis_df = pd.DataFrame(kurtosis(data))
    kurtosis_df['Fault'] = fault
    kurtosis_list.append(kurtosis_df)
kurtosis_all = pd.concat(kurtosis_list, axis=0, ignore_index=True)

#Mean
mean_list = []
for fault in faults:
    mean_df = mean(eval(fault))
    mean_df['Fault'] = fault
    mean_list.append(mean_df)
mean_all = pd.concat(mean_list, axis=0, ignore_index=True)

#Standard Deviation
std_list = []
for fault in faults:
    std_df = std(eval(fault))
    std_df['Fault'] = fault
    std_list.append(std_df)
std_all = pd.concat(std_list, axis=0, ignore_index=True)

#Skewness
skew_list = []
for fault in faults:
    data = eval(fault)
    skew_df = pd.DataFrame(skew(data))
    skew_df['Fault'] = fault
    skew_list.append(skew_df)
skew_all = pd.concat(skew_list, axis=0, ignore_index=True)

#Impulse Factor
Impulsef_list = []
for fault in faults:
    Impulsef_df = Impulsef(eval(fault))
    Impulsef_df['Fault'] = fault
    Impulsef_list.append(Impulsef_df)
Impulsef_all = pd.concat(Impulsef_list, axis=0, ignore_index=True)

#Crest Factor
crestf_list = []
for fault in faults:
    crestf_df = crestf(eval(fault))
    crestf_df['Fault'] = fault
    crestf_list.append(crestf_df)
crestf_all = pd.concat(crestf_list, axis=0, ignore_index=True)

#Variance
Variance_list = []
for fault in faults:
    Variance_df = Variance(eval(fault))
    Variance_df['Fault'] = fault
    Variance_list.append(Variance_df)
Variance_all = pd.concat(Variance_list, axis=0, ignore_index=True)

#Clearance Factor
clearancef_list = []
for fault in faults:
    clearancef_df = clearancef(eval(fault))
    clearancef_df['Fault'] = fault
    clearancef_list.append(clearancef_df)
clearancef_all = pd.concat(clearancef_list, axis=0, ignore_index=True)

#Square Root Amplitude
SQRT_AMPL_list = []
for fault in faults:
    SQRT_AMPL_df = SQRT_AMPL(eval(fault))
    SQRT_AMPL_df['Fault'] = fault
    SQRT_AMPL_list.append(SQRT_AMPL_df)
SQRT_AMPL_all = pd.concat(SQRT_AMPL_list, axis=0, ignore_index=True)

#Form Factor
ff_list = []
for fault in faults:
    ff_df = ff(eval(fault))
    ff_df['Fault'] = fault
    ff_list.append(ff_df)
ff_all = pd.concat(ff_list, axis=0, ignore_index=True)
```
